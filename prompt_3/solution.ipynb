```python
#Solution 1
class LinkedChar:
    def __init__(self, char_mat, char: str, row: int, col: int):
        self.char_mat: list[list[LinkedChar]] = char_mat
        self.char = char
        self.row = row
        self.col = col
    
    def __repr__(self) -> str:
        return f'({self.char}, {self.row}, {self.col})'
    
    def look_for_surrounding_printable(self, ignore_chars, ignore_numeric = True):
        for row in [-1, 0, 1]:
            for col in [-1, 0, 1]:
                if row == 0 == col:
                    continue
                try:
                    linked_char : LinkedChar = self.char_mat[self.row + row][self.col + col]
                except:
                    continue
                if linked_char.char in ignore_chars or (ignore_numeric and linked_char.char.isdigit()):
                    continue
                if linked_char.char.isprintable():
                    return True
        return False
                
        
class LinkedCharMatrix:
    def __init__(self):
        self.matrix : list[list[LinkedChar]]= []
    
    @property
    def insert_row_pos(self):
        return len(self.matrix) - 1
    
    @property
    def insert_column_pos(self):
        return len(self.matrix[self.insert_row_pos])
    
    @property
    def row_count(self):
        return len(self.matrix)
    
    @property
    def column_count(self):
        return len(self.matrix[self.insert_row_pos])

    def add_row(self):
        self.matrix.append([])
    
    def add_linked_char(self, item):
        self.matrix[self.insert_row_pos].append(LinkedChar(self.matrix, item, self.insert_row_pos, self.insert_column_pos))
        
    def find_part_numbers(self):
        num_list = []
        curr_num = ''
        validated = False
        for row in self.matrix:
            for linked_char in row:
                if linked_char.char.isnumeric():
                    curr_num += linked_char.char
                    if not validated:
                        validated = linked_char.look_for_surrounding_printable('.')
                else:
                    if validated:
                        num_list.append(int(curr_num))
                    validated = False
                    curr_num = ''
        return num_list
    
input_file = 'input.txt'
total = 0
matrix = LinkedCharMatrix()
with open(input_file) as file:
    for line in file:
        matrix.add_row()
        for char in line:
            if char == '\n':
                continue
            matrix.add_linked_char(char)
            
sum(matrix.find_part_numbers())

```

```python
#Solution 2
class LinkedChar:
    def __init__(self, char_mat, char: str, row: int, col: int):
        self.char_mat: list[list[LinkedChar]] = char_mat
        self.char = char
        self.row = row
        self.col = col
    
    def __repr__(self) -> str:
        return f'({self.char}, {self.row}, {self.col})'
    
    def get_next(self):
        return self.char_mat[self.row][self.col+1]
        
    def get_prev(self):
        return self.char_mat[self.row][self.col-1]
    
    def if_me_number_make_whole(self, reference: list):
        '''
        Refernce exists so that if an object shows up in this 
        funct that is in the reference is can be removed
        '''
        full_num = self.char
        prev = self.get_prev()
        while prev.char.isnumeric():
            if prev in reference:
                reference.remove(prev)
            full_num = prev.char + full_num
            prev : LinkedChar = prev.get_prev()
        next = self.get_next()
        while next.char.isnumeric():
            if next in reference:
                reference.remove(next)
            full_num = full_num + next.char
            try:
                next : LinkedChar = next.get_next()
            except:
                break
        return int(full_num)
    
    def find_for_surrounding_numbers(self) -> list[int]:
        nums : list[LinkedChar] = []
        for row in [-1, 0, 1]:
            for col in [-1, 0, 1]:
                linked_char : LinkedChar = self.char_mat[self.row + row][self.col + col]
                if linked_char.char.isnumeric():
                    nums.append(linked_char)
        full_nums = []
        while len(nums):
            nu = nums.pop()
            full_nums.append(nu.if_me_number_make_whole(nums))
            
        return full_nums
        
class LinkedCharMatrix:
    def __init__(self):
        self.matrix : list[list[LinkedChar]]= []
    
    @property
    def insert_row_pos(self):
        return len(self.matrix) - 1
    
    @property
    def insert_column_pos(self):
        return len(self.matrix[self.insert_row_pos])
    
    @property
    def row_count(self):
        return len(self.matrix)
    
    @property
    def column_count(self):
        return len(self.matrix[self.insert_row_pos])

    def add_row(self):
        self.matrix.append([])
    
    def add_linked_char(self, item):
        self.matrix[self.insert_row_pos].append(LinkedChar(self.matrix, item, self.insert_row_pos, self.insert_column_pos))
        
    def find_gears(self):
        num_list = []
        for row in self.matrix:
            for linked_char in row:
                if linked_char.char == '*':
                    nums = linked_char.find_for_surrounding_numbers()
                    if len(nums) == 2:
                        num_list.append(nums[0] * nums[1])
        return num_list
    
    
input_file = 'input.txt'
total = 0
matrix = LinkedCharMatrix()
with open(input_file) as file:
    for line in file:
        matrix.add_row()
        for char in line:
            if char == '\n':
                continue
            matrix.add_linked_char(char)     
sum(matrix.find_gears())
```
