```python
# Solution 2
class RangeDict(dict):
    def get_in_range(self, key):
        delta = 0
        if type(key) is not type(range):
            for i_key in self.keys():
                if key in i_key:
                    delta = key - i_key.start
                    key = i_key
                    return [super().__getitem__(key), delta]
            return [key, delta]
        return [super().__getitem__(key), delta]
    
    def sorted_keys(self) -> list[range]:
        return sorted(self.keys(), key=lambda r: r.start)
        
    def get_ranges_of_in_range(self, rang: range):
        ret_ranges = []
        iterrange = self.sorted_keys()
        
        if rang.start < iterrange[0].start:
            if rang.stop > iterrange[0].start:
                ret_ranges.append(range(rang.start, iterrange[0].start))
                rang = range(iterrange[0].start, rang.stop)
            else:
                ret_ranges.append(rang)
                return ret_ranges
            
        for in_range in iterrange:
            if rang.start < in_range.start:
                rang = range(in_range.start, rang.stop)
            if rang.start < in_range.stop:
                if rang.stop >= in_range.stop:
                    start = sum(self.get_in_range(rang.start))
                    stop  = sum(self.get_in_range(in_range.stop-1))
                    ret_ranges.append(range(start, stop))
                else:
                    start = sum(self.get_in_range(rang.start))
                    stop  = sum(self.get_in_range(rang.stop-1))
                    ret_ranges.append(range(start, stop))
                    return ret_ranges
                    
        return ret_ranges
            
maps: list[RangeDict] = [RangeDict(),
        RangeDict(),
        RangeDict(),
        RangeDict(),
        RangeDict(),
        RangeDict(),
        RangeDict()]

def access_maps(index, accessors: list[range]):
    try:
        next_accessors : list[range] = []
        for accessor in accessors:
            next_accessors.extend(maps[index].get_ranges_of_in_range(accessor))
    except IndexError:
        return accessors
    return access_maps(index+1, next_accessors)

input_file = 'input.txt'
pos = -1
first = True
seed_ranges = []

with open(input_file) as file:
    for line in file:
        if first:
            new = line.split(': ')[1]
            new = new.split()
            seed_ranges = [range(int(new[i]), int(new[i]) + int(new[i+1])) for i in range(0, len(new), 2)]
            first = False
        elif line[0].isprintable() and not line[0].isnumeric():
            pos += 1
        elif line[0].isnumeric():
            new = line.split()
            new = [int(v) for v in new]
            maps[pos][range(new[1], new[1] + new[2])] = new[0]
locs = []
for i, seed in enumerate(seed_ranges):
    val =  access_maps(0,[seed])
    locs.extend(val)
sorted(locs, key=lambda r: r.start)[0].start
            
```
